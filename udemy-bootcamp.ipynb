```

```



````python
reg.score(x,y)
````

 membuat adjusted r^2 rumusnya adalah



```
# adjusted_r2 = 1-(1-r2) * (n-1) / (n-p-1)
```



n = jumlah total colum yang di gunakan


p = jumlah data row yang digunakan


r2 = adalah var.score(x,y)

pada contoh kali ini x disini ada SAT dan Rand 1,2,3 jadi nilai p = 2 dan total columnya masing masing adalah 84 jadi n = 84

untuk mengetahui nilai n dan p bisa menggunakan 


```
# var.shape
```
pada contoh ini x.shape hasilnya adalah (84, 2) sama seperti data di atas

jadi ketika ingin memasukan datanya bisa seperi ini

```
# n = 84
# p = 2

atau lebih mudahnya bisa seperti ini 

# n = x.shape[0] 
# p = x.shape[1]
```









````python
x.shape
````

````python
r2 = reg.score(x,y)

n = x.shape[0]
p = x.shape[1]

adjusted_r2 = 1-(1-r2) * (n-1) / (n-p-1)
adjusted_r2
````

pengertianya adalah mengubah data menjadi skala standar caranya adalah mengurangi mean dan membaginya dengan standar rumusnya adalah

standarized fariable = (x-u) / o

x = original variable

u =original variable

o = standar deviation dari original variable
